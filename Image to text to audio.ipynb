```python
import pytesseract
from PIL import Image
import gtts
```

```python
img = Image.open('/content/text1.png')
print(img)
```

```python
result = pytesseract.image_to_string(img)
print(result)
tts = gtts.gTTS(result)
tts.save("hello.mp3")
```

```python
from gtts import gTTS
from IPython.display import Audio
```
